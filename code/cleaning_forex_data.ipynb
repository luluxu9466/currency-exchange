```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# csv directory
file = '../data/forex_dta.csv'

# read csv
df = pd.read_csv(file)

# list of months
months = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "August", "Sept", "Oct", "Nov", "Dec"]

# slice the "Rate" column and create new columns for the currency (three letter initials) 
# and current rate (with relation to EUR)
df["currency"] = df["Rate"].str.slice(2, 5, 1) 
df["rate"] = df["Rate"].str.slice(8,-1, 1)

# slice date and create columns for year and month
df["year"] = df["Date"].str.slice(0, 4, 1) 
df["month"] = df["Date"].str.slice(5,7, 1)
df["month"] = pd.to_numeric(df["month"])
df["month_str"] = ""

for i in range(480):
    df["month_str"][i] = months[df["month"][i] - 1]
```
